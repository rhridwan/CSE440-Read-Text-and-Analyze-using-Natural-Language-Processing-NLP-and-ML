```python
import nltk
from bs4 import BeautifulSoup
import urllib.request
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
import string

print("Reading Text1 and Tokenizing...")
response = urllib.request.urlopen('https://www.gutenberg.org/files/74/74-h/74-h.htm')
html = response.read()
print("Reading Text2 and Tokenizing...")
response1 = urllib.request.urlopen('https://www.gutenberg.org/files/32325/32325-h/32325-h.htm')
html1 = response1.read()
```

```python
nltk.download("all")
```

```python
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
print(text)
```

```python
soup1 = BeautifulSoup(html1,"html5lib")
text1 = soup1.get_text(strip=True)
print(text1)
```

```python
tokens = [t for t in text.split()]
print("Removing Punctuation frist html text...")
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text)
tokens = [t.lower() for t in tokens]
print(tokens)
```

```python
tokens1 = [t1 for t1 in text1.split()]
print("Removing Punctuation second html txt...")
tokenizer1 = RegexpTokenizer(r'\w+')
tokens1 = tokenizer1.tokenize(text1)
tokens1 = [t1.lower() for t1 in tokens1]
print(tokens1)
```

```python
print("Removing Stop Words from first htm txt...")
clean_tokens = tokens[:]
sr = stopwords.words('english')
for token in tokens:
	if token in stopwords.words('english'):
		clean_tokens.remove(token)
print(clean_tokens)
```

```python
print("Removing Stop Words from second html txt ...")
clean_tokens1 = tokens1[:]
sr1 = stopwords.words('english')
for token1 in tokens1:
	if token1 in stopwords.words('english'):
		clean_tokens1.remove(token1)
print(clean_tokens1)
```

```python
print("Frequency Analysis first html txt...")
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
	print (str(key) + ':' + str(val))
print("Length of Unique Items:", len(freq.items()))
freq.plot(20, cumulative=False)
```

```python
print("Frequency Analysis second html txt...")
freq1 = nltk.FreqDist(clean_tokens1)
for key1,val1 in freq1.items():
	print (str(key1) + ':' + str(val1))
print("Length of Unique Items:", len(freq1.items()))
freq1.plot(20, cumulative=False)
```

```python
print("Frist html file Stemming...")
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in clean_tokens]
print("Frist html file Frequency Analysis for Stemming...")
freq = nltk.FreqDist(stemmed_tokens)
for key,val in freq.items():
	print (str(key) + ':' + str(val))
print("Length of Unique Items:", len(freq.items()))
freq.plot(20, cumulative=False)
```

```python
print("Second html file Stemming...")
stemmer1 = PorterStemmer()
stemmed_tokens1 = [stemmer1.stem(token1) for token1 in clean_tokens1]
print("Second html file  Frequency Analysis for stemming...")
freq1 = nltk.FreqDist(stemmed_tokens1)
for key1,val1 in freq1.items():
	print (str(key1) + ':' + str(val1))
print("Length of Unique Items:", len(freq1.items()))
freq1.plot(20, cumulative=False)
```

```python
print("first html file Lemmatizing...")

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in stemmed_tokens]
print("Frist html Frequency Analysis...")
freq = nltk.FreqDist(lemmatized_tokens)

for key,val in freq.items():
	print (str(key) + ':' + str(val))
print("Length of Unique Items:", len(freq.items()))
freq.plot(20, cumulative=False)
```

```python
print("Second html file Lemmatizing...")

from nltk.stem import WordNetLemmatizer
lemmatizer1 = WordNetLemmatizer()
lemmatized_tokens1 = [lemmatizer1.lemmatize(token1) for token1 in stemmed_tokens1]
print("Second html Frequency Analysis...")
freq1 = nltk.FreqDist(lemmatized_tokens1)

for key1,val1 in freq1.items():
	print (str(key1) + ':' + str(val1))
print("Length of Unique Items:", len(freq1.items()))
freq1.plot(20, cumulative=False)
```

```python
print("First html txt POS Analysis...")
import operator
import matplotlib.pyplot as plt
pos = nltk.pos_tag(lemmatized_tokens)
pos_counts = {}
for key,val in pos:
	print(str(key) + ':' + str(val))
	if val not in pos_counts.keys():
		pos_counts[val] = 1
	else:
		pos_counts[val] += 1
print(pos_counts)
plt.bar(range(len(pos_counts)), list(pos_counts.values()), align='center')
plt.xticks(range(len(pos_counts)), list(pos_counts.keys()))
plt.show()
```

```python
print("Second html txt POS Analysis...")
import operator
import matplotlib.pyplot as plt
pos1 = nltk.pos_tag(lemmatized_tokens1)
pos_counts1 = {}
for key1,val1 in pos1:
	print(str(key1) + ':' + str(val1))
	if val1 not in pos_counts1.keys():
		pos_counts1[val1] = 1
	else:
		pos_counts1[val1] += 1
print(pos_counts1)
plt.bar(range(len(pos_counts1)), list(pos_counts1.values()), align='center')
plt.xticks(range(len(pos_counts1)), list(pos_counts1.keys()))
plt.show()
```
